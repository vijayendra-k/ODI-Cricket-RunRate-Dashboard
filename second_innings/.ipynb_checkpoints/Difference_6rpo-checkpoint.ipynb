```python
import pandas as pd
```

```python
df1=pd.read_csv('above6.csv')
```

```python
df2=pd.read_csv('below6.csv')
```

```python
df1=df1.rename(columns={'rpo':'rpo_rrr more than 6'})
df2=df2.rename(columns={'rpo':'rpo_rrr less than or equals 6'})
```

```python
df1=df1[df1['runs']>=500]
```

```python
df2=df2[df2['runs']>=500]
```

```python
result=pd.merge(df1,df2,how='outer',on='bat')
```

```python
result
```

```python
result=result[['bat','rpo_rrr less than or equals 6','rpo_rrr more than 6']]
```

```python
result
```

```python
result.to_csv('Difference_6rpo.csv')
```
