```python
bowlunder6=dataset[(dataset['inns_rr']>5) & (dataset['inns_rr']<=6)]
```

```python
bowlunder6.columns
```

```python
dataset['dismissal'].value_counts()
```

```python
bowldis=['caught','bowled','leg before wicket','stumped','hit wicket']
bowlunder6['wickets_earned?']=bowlunder6['dismissal'].apply(lambda x: 1 if x in bowldis else 0)
```

```python
bowlunder6['false_shot']=1-bowlunder6['control']
req_cols=['bowl','bowlruns','ballfaced','wickets_earned?','false_shot']
bowlunder6=bowlunder6[req_cols]
```

```python
bowlunder6
```

```python
ans=bowlunder6.groupby('bowl').sum().reset_index().sort_values(by='wickets_earned?',ascending=False)
```

```python
ans
```

```python
ans['bowl_avg']=ans['bowlruns']/ans['wickets_earned?']
ans['bowl_avg']=round(ans['bowl_avg'],2)
ans['balls_per_wicket']=ans['ballfaced']/ans['wickets_earned?']
ans['balls_per_wicket']=round(ans['balls_per_wicket'],2)
ans['false_shot%']=ans['false_shot']*100/ans['ballfaced']
ans['false_shot%']=round(ans['false_shot%'],2)
ans['econ']=ans['bowlruns']*6/ans['ballfaced']
ans['econ']=round(ans['econ'],2)
```

```python
ans=ans.sort_values(by='econ',ascending=True)
```

```python
ans=ans.rename(mapper={'bowl':'bowler','bowlruns':'runs_conceded','ballfaced':'balls_bowled','wickets_earned?':'wickets'},axis=1)
ans=ans.drop(columns=['false_shot'])
```

```python
totalballs=pd.read_csv('totalballs.csv')
totalballs.drop('Unnamed: 0',axis=1,inplace=True)
totalballs=totalballs.rename(mapper={'bowl':'bowler'},axis=1)
final=pd.merge(ans,totalballs,on='bowler')
final['%career_balls']=final['balls_bowled']*100/final['ballfaced']
final['%career_balls']=round(final['%career_balls'],2)
final.drop(columns=['ballfaced'],axis=1,inplace=True)
```

```python
final.to_csv('bowl_under6.csv')
```
