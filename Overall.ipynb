Imported the pandas library which I am using for Data Manipulation in this project.

```python
import pandas as pd
```

Read the dataset using the Pandas read_csv feature

```python
dataset=pd.read_csv('bbb_odi.csv')
```

```python
# dataset.groupby(['bat'])
dataset
```

Create a dictionary where you can store all the key pair values of the batter and bowler, which u will use later for finding who got out.

```python
dictbat={}
for i,x in dataset.iterrows():
#     print(x)
#     break
    if x['p_bat'] not in dictbat:
        dictbat[x['p_bat']]=x['bat']
    if x['p_bowl'] not in dictbat:
        dictbat[x['p_bowl']]=x['bowl']
```

```python
dictbat
```

Adding all the modes of dismissals which I will use to judge whether a batter got out or not.

```python
len(dictbat)
modes=['caught','bowled','leg before wicket','run out','stumped','hit wicket','obstructing the field','hit wicket','handled the ball']
```

A new column who got out has to be created in the dataset to find out which batter got out.

```python
dataset['who_out?']=''
```

Using the key pair value from the dictionary where we stored the data, we need to find the batter's name who has been dismissed instead of using the 'id'

```python
for index, row in dataset.iterrows():
    if row['dismissal'] in modes:
        dataset.at[index, 'who_out?'] = dictbat[row['p_out']]
```
